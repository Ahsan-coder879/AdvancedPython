```python
def calculate_tax(amount, rate):
    """
    Calculates the tax amount based on a price and a percentage rate.
    :param amount: The total price (float/int)
    :param rate: The tax rate as a decimal (e.g., 0.07 for 7%)
    :return: The calculated tax value
    """
    tax_total = amount * rate
    return tax_total

price_a = 150.00
tax_rate_a = 0.07  # 7%
sales_tax = calculate_tax(price_a, tax_rate_a)

price_b = 12000.00
tax_rate_b = 0.15
luxury_tax = calculate_tax(price_b, tax_rate_b)

print(f"Standard Item ($150): Tax = ${sales_tax:.2f}")
print(f"Luxury Item ($12,000): Tax = ${luxury_tax:.2f}")

print(f"Small Purchase ($10): Tax = ${calculate_tax(10, 0.05):.2f}")
```
