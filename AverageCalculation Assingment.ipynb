```python
print("======Marks & Grade System=====")
numSubjects = int(input("How many subjects?:  "))
total_marks = 0
for i in range(numSubjects):
    name = input("Enter subject name: ")
    mark = float(input(f"Enter marks for {name}: "))
    total_marks = total_marks + mark
average = total_marks / numSubjects
if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"
print("\n--- Results ---")
print("Total Marks: ", total_marks)
print("Average:     ", average)
print("Final Grade: ", grade)
```
