```python
class pclass:

    def parentClassFunction(self):
        print("this function is from parent class")

class cClass(pclass):

    def childClassFunction (self):
        print("this function is from child class")

Child = cClass()
Child.parentClassFunction()
Child.childClassFunction()
```

```python
class parentClass:
    def __init__(self,_name):
        self.__name = _name

    def publicFunc(self):
        print("this function is from parent class")
    def privateFunc(self):
        print("this function is from parent class")

class childClass(parentClass):
    def __init__(self,name,age):
        super().__init__(name)
        self.age = age

    def chilClassFunc(self):
        print("this function is from child class")
class grandChildClass(childClass):
    def __init__(self,name,age):
        super().__init__(name,age)
        self.age = age
```

```python
ca = grandChildClass("Ahmad",20)
ca.privateFunc()
```

```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd

response = r.get("https://www.scrapethissite.com/pages/simple/")
pageSource = response.text
with open("PageSource.txt","w",encoding="utf-8") as f:
  f.write(pageSource)
```

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
with open("PageSource.txt","r",encoding="utf-8") as f:
    sourceCode = f.read()
    s = bs(sourceCode,"html.parser")
    # s = s.prettify()
    countryData ={"Country Name": [], "Country Capital": [], "Country Population": [], "Country Area": []}
    countryNames = s.find_all("h3",{"class":"country-name"})
    countrycapitals = s.find_all("span", {"class": "country-capital"})
    countrypopulation = s.find_all("span", {"class": "country-population"})
    countryArea = s.find_all("span", {"class": "country-area"})
    for i in range(len(countryNames)):
        countryData["Country Name"].append(countryNames[i].text.strip())
        countryData["Country Capital"].append(countrycapitals[i].text.strip())
        countryData["Country Population"].append(countrypopulation[i].text.strip())
        countryData["Country Area"].append(countryArea[i].text.strip())
```

```python
countryInfodf = pd.DataFrame(countryData)
```

```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd

url = "https://www.deckshop.pro/card/list"
response = r.get(url)
soup = bs(response.text, "html.parser")

cards = soup.find_all("img", {"class": "card"})
cardData = {"Card Name": [], "Card Url": []}
for card in cards:
    name = card.get("alt")
    img_url = card.get("src")

    if name and img_url:
        cardData["Card Name"].append(name.strip())
        cardData["Card Url"].append(img_url.strip())

cardInfodf = pd.DataFrame(cardData)
cardInfodf.to_csv("ClashRoyalCards.csv", index=False)
print(cardInfodf.head())
```
