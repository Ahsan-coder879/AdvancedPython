```python
# check by comparing id from boths sheets and see if all the products from
#Collections sheet exist in items sheet
# Merge both sheets
# Compare both ids = On id's getting matched retrieve the information
# For the specific product Id, Name , Secondary Name , barcode , price from collection
# for same product you can get description , secondary Description, Menu , Brand and Secondary brand from items sheet
# Attach new row at the end of items sheet
```

```python
file = "VendorProducts.xlsx"
```

```python
dfItems = pd.read_excel(file,sheet_name="Items")
dfCollections = pd.read_excel(file,sheet_name="Collections")
```

```python
matchIds = dfCollections["Item Id"].isin(dfItems["Id"])
```

```python
existsIds = matchIds.all()
```

```python
if existsIds:
    print("All collection sheet Item id exist in the Items id Sheet.")
else:
    missing_products = dfCollections[~matchIds]
    print(f"Missing products found: {len(missing_products)}")
    print(missing_products)
```

```python
dfCollections
```

```python
CollectoinSheetIds = dfCollections["Item Id"]
CollectoinSheetIds
```

```python
ItemSheetIds =  dfItems["Id"].tolist()
```

```python
ItemSheetIds
for eachId in CollectoinSheetIds:
    if eachId in ItemSheetIds:
        picture  = dfCollections.loc[dfCollections["Item Id"]==eachId,"Picture"]
        Barcode  = dfCollections.loc[dfCollections["Item Id"]==eachId,"Barcode"]
        Item_Name  = dfCollections.loc[dfCollections["Item Id"]==eachId,"Item Name"]
        Secondary_Item_Name  = dfCollections.loc[dfCollections["Item Id"]==eachId,"Secondary Item Name"]
        Price  = dfCollections.loc[dfCollections["Item Id"]==eachId,"Price"]

        Description  = dfItems.loc[dfItems["Id"]==eachId,"Description"]
        Secondary_Description  = dfItems.loc[dfItems["Id"]==eachId,"Secondary Description"]
        Menu  = dfItems.loc[dfItems["Id"]==eachId,"Menu"]
        Brand  = dfItems.loc[dfItems["Id"]==eachId,"Brand"]
        Secondary_Brand  = dfItems.loc[dfItems["Id"]==eachId,"Secondary Brand"]
        currentrowNumber = len(dfItems["Id"])
        dfItems.loc[currentrowNumber] = [picture,eachId,Item_Name,Secondary_Item_Name,Barcode,"NAN",Price,Description,Secondary_Description,Menu,Brand,Secondary_Brand,False]
        print("Matched"+" "+str(eachId)+" "+str(currentrowNumber))

    else:
        print("Not Matched")
```
