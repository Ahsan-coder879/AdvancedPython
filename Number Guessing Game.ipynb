```python
import random

generatedNumber = random.randint(1, 100)
attempts = 0
guess = 0
print("The generated number is between 1 to 100.")
while guess != generatedNumber:
    guess = int(input("Enter your guess: "))
    attempts = attempts + 1
    if guess > generatedNumber:
        print("Too High")
    elif guess < generatedNumber:
        print("Too Low")
    else:
        print("Correct!")
        print(f"You guessed the number in {str(attempts)} attempts.")
```
