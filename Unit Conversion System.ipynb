```python
print("======Unit Conversion System======")

while True:
    print("\nSelect From a Category:")
    print("1. Length (m to km OR in to ft)")
    print("2. Temperature (C to F)")
    print("3. Weight (kg to g OR kg to lb)")
    print("4. Time (sec to min OR sec to hr)")
    print("0. Exit")
    choice = input("\nEnter choice (0-4): ")
    if choice == "0":
        print("Exit!")
        break
    if choice == "1":
        print("A. Meters to Kilometers")
        print("B. Inches to Feet")
        subCategory = input("Select A or B: ").upper()
        if subCategory == "A":
            meters = float(input("Enter Meters: "))
            print(f"{meters}, meter = {meters / 1000} km")
        elif subCategory == "B":
            inches = float(input("Enter Inches: "))
            print(inches, "in =", inches / 12, "ft")
    elif choice == "2":
        val = float(input("Enter Celsius: "))
        print(val, "°C =", (val * 9 / 5) + 32, "°F")
    elif choice == "3":
        print("A. Kilograms to Grams")
        print("B. Kilograms to Pounds")
        subCategory = input("Select A or B: ").upper()
        if subCategory == "A":
            kilograms = float(input("Enter Kg: "))
            print(kilograms, "kg =", kilograms * 1000, "g")
        elif subCategory == "B":
            kilograms = float(input("Enter Kg: "))
            print(kilograms, "kg =", kilograms * 2.20462, "lbs")
    elif choice == "4":
        print("A. Seconds to Minutes")
        print("B. Seconds to Hours")
        subCategory = input("Select A or B: ").upper()
        if subCategory == "A":
            seconds = float(input("Enter Seconds: "))
            print(seconds, "sec =",seconds / 60, "min")
        elif subCategory == "B":
            seconds = float(input("Enter Seconds: "))
            print(seconds, "sec =", seconds / 3600, "hr")
    else:
        print("Invalid category selection.")
```
